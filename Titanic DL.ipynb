```python
import pandas as pd
```

```python
titanic_train = pd.read_csv("E:\\MyCodes\\Titanic\\train.csv")
```

```python
data = titanic_train.copy()
data.isnull().sum()
```

```python
data = data.drop(columns=['Cabin','Name','Ticket'])
```

```python
data = data.dropna(subset=['Embarked'])
```

```python
embrkd = pd.get_dummies(data['Embarked'])
pcl = pd.get_dummies(data['Pclass'])
sex =  pd.get_dummies(data['Sex'])
```

```python
data = pd.concat([data,embrkd,pcl,sex],axis=1)
```

```python
data = data.drop(columns=['Pclass','Embarked','Sex'])
data
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
data.isnull().sum()
```

```python
data
```

```python
x = data.drop(columns=['PassengerId','Survived'])
y = data['Survived']
```

```python
from tensorflow import keras
from tensorflow.keras.callbacks import Callback
           
faremodel=keras.Sequential([
                        keras.layers.Dense(256,activation='relu'),
                        keras.layers.Dropout(.4),
                        keras.layers.Dense(128,activation='relu'),
                        keras.layers.Dense(1,activation='sigmoid')
])

faremodel.compile(
               optimizer='adam',
               loss='mean_absolute_error',
               metrics=['accuracy']
)

faremodel.fit(x,y,epochs=3000,batch_size=500,verbose=2,validation_split=.3)

#print("EVALUATE")
#model.evaluate(test_images,test_labels)
```
