```python
import pandas as pd
```

```python
titanic_train = pd.read_csv("E:\\MyCodes\\Titanic\\train.csv")
titanic_test = pd.read_csv("E:\\MyCodes\\Titanic\\test.csv")
titanic_test.isnull().sum()
```

```python
titanic_test[:][0:1]
```

```python
data = titanic_train.copy()
for i in range(len(titanic_test)):
    data = data.append(titanic_test[:][i:i+1])
data
```

```python
data = data.drop(columns=['Cabin','Name','Ticket'])
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=['Embarked'])
data.isnull().sum()
```

```python
embrkd = pd.get_dummies(data['Embarked'])
pcl = pd.get_dummies(data['Pclass'])
sex =  pd.get_dummies(data['Sex'])
```

```python
data = pd.concat([data,embrkd,pcl,sex],axis=1)
data
```

```python
data = data.drop(columns=['Pclass','Embarked','Sex'])
data
```

```python
data = data.reset_index(drop=True)
data
```

```python
data.isnull().sum()
```

```python
data_clean = data.copy()
data_clean = data_clean.drop(columns='Survived')
data_clean = data_clean.dropna(subset=['Age','Fare'])
data_clean.isnull().sum()
```

```python
data_clean
```

```python
x = data_clean.drop(columns=['PassengerId','Fare'])
y = data_clean['Fare']
```

```python
from tensorflow import keras
from tensorflow.keras.callbacks import Callback


class NBatchLogger(Callback):
    def __init__(self, display):
        self.seen = 0
        self.display = display

    def on_batch_end(self, batch, logs={}):
        self.seen += logs.get('size', 0)
        if self.seen % self.display == 0:
            # you can access loss, accuracy in self.params['metrics']
            print('\n{}/{} - loss ....\n'.format(self.seen, self.params['mean_absolute_error'])) 
           
faremodel=keras.Sequential([
                        keras.layers.Dense(32,activation='relu'),
                        keras.layers.Dropout(.2),
                        keras.layers.Dense(64,activation='relu'),
                        keras.layers.Dense(1)
])

faremodel.compile(
               optimizer='adam',
               loss='mean_absolute_error',
               metrics=['accuracy']
)

#faremodel.fit(x,y,epochs=30000,batch_size=500,verbose=2)

out_batch = NBatchLogger(display=100)
faremodel.fit(x,y,batch_size=521,epochs=30000,callbacks=[out_batch])

#print("EVALUATE")
#model.evaluate(test_images,test_labels)
```
